# Linear Regression - Predicting Per Capita Income

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

### Displaying dataset

```python
data = pd.read_csv('C:\\Users\\KAAVIYA\\OneDrive\\Desktop\\Data\\income.csv')
data.head()
```

### Plotting the graph

```python
%matplotlib inline
plt.title('Year vs Income')
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(data.year,data.income,color='red',marker='+')
```

### Creating an object of LinearRegression class

```python
reg = linear_model.LinearRegression()
```

### Training the model

```python
data['year'] = pd.to_numeric(data['year'], errors='coerce')
data['income'] = pd.to_numeric(data['income'], errors='coerce')
```

```python
reg.fit(data[['year']], data[['income']])
```

### Predicting the per capita income in 2022

```python
pred = reg.predict([[2022]])
pred
```

### Plotting the graph

```python
%matplotlib inline
plt.title('Year vs Income')
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(data.year,data.income,color='red',marker='+', label = 'Data Points')
plt.plot(data.year,reg.predict(data[['year']]),color='blue', label = 'Line of Best Fit')
plt.legend()
plt.show
```
